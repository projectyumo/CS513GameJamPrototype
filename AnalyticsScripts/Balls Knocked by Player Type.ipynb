```python
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

# Load the data
filename = 'lastcallstudios-c8991-default-rtdb-export.json'
data = pd.read_json(filename)

# Display the head of the dataframe
data.head()
```

```python
# Extract the 'analytics_v5' column
analytics_v5 = data['analytics_v5'].apply(pd.Series)

# Display the head of the dataframe
analytics_v5.head()
```

```python
# Filter out rows where 'currentLevel' is 0 or -1
analytics_v5_filtered = analytics_v5[(analytics_v5['currentLevel'] != 0) & (analytics_v5['currentLevel'] != -1)]

# Group by 'currentLevel' and calculate the sum of 'ballsKnockedByGhost' and 'ballsKnockedByPlayer'
grouped_data = analytics_v5_filtered.groupby('currentLevel')[['ballsKnockedByGhost', 'ballsKnockedByPlayer']].sum().reset_index()

# Display the head of the grouped data
grouped_data.head()
```

```python
# Filter out rows where 'currentLevel' is 0 or -1
analytics_v5_filtered = analytics_v5[(analytics_v5['currentLevel'] != 0) & (analytics_v5['currentLevel'] != -1)]

# Group by 'currentLevel' and calculate the sum of 'ballsKnockedByGhost' and 'ballsKnockedByPlayer'
grouped_data = analytics_v5_filtered.groupby('currentLevel')[['ballsKnockedByGhost', 'ballsKnockedByPlayer']].sum().reset_index()

# Display the head of the grouped data
grouped_data.head()
```

```python
grouped_data['currentLevel'] = grouped_data['currentLevel'].astype(int)
grouped_data.head()
```

```python
# Create a bar plot
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='currentLevel', y='value', hue='variable', data=pd.melt(grouped_data, ['currentLevel']), palette='viridis')

# Set the labels and title
ax.set(xlabel='Current Level', ylabel='Number of Balls')
plt.title('Number of Balls Knocked by Ghost and Player for Each Level')

# Change the legend labels
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, ['Balls Knocked by Ghost', 'Balls Knocked by Player'], title='Category', title_fontsize='13')

plt.show()
```
